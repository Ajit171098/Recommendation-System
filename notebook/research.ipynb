```python
ratings = pd.read_csv('Ratings.csv', sep=",", on_bad_lines='skip', encoding='latin-1')
ratings.head()
```

```python
ratings.shape
```

```python
ratings.rename(columns={'User-ID': 'user_id', 'Book-Rating': 'rating'}, inplace=True)
```

```python
ratings.shape
```

```python
users = pd.read_csv('Users.csv', sep=",", on_bad_lines='skip', encoding='latin-1')
users.head()
```

```python
users.rename(columns={'User-ID': 'user_id', 'Location': 'location', 'Age': 'age'}, inplace=True)
users.head()
```

```python
users.shape
```

```python
no_of_books_read_by_userid = ratings['user_id'].value_counts() > 200
no_of_books_read_by_userid
```

```python
user_id_greater_than_200 = no_of_books_read_by_userid[no_of_books_read_by_userid].index
user_id_greater_than_200
```

```python
users_who_rated_morethan_200times = ratings[ratings['user_id'].isin(user_id_greater_than_200)]
users_who_rated_morethan_200times
```

```python
ratings_with_books = books.merge(users_who_rated_morethan_200times, on='ISBN')
ratings_with_books
```

```python
number_of_ratings_per_book = ratings_with_books.groupby('book_title')['rating'].count().reset_index()
number_of_ratings_per_book
```

```python
number_of_ratings_per_book.rename(columns={'rating': 'number_of_ratings'}, inplace=True)
number_of_ratings_per_book

```

```python
final_books = number_of_ratings_per_book.merge(ratings_with_books, on='book_title')
final_books
```

```python
final_books.shape
```

```python
final_books = final_books[final_books['number_of_ratings'] >= 50]
final_books.shape
```

```python
final_books
```

```python
final_books.drop_duplicates(subset=['user_id', 'book_title'], inplace=True)
final_books.shape
```

```python
final_books
```

```python
final_books.isnull().sum()
```

```python
book_pivot = final_books.pivot_table(index='book_title', columns='user_id', values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
%pip install scipy
```

```python
from scipy.sparse import csr_matrix

book_sparse_matrix = csr_matrix(book_pivot)
book_sparse_matrix
```

```python
type(book_sparse_matrix)
```

```python
%pip install scikit-learn
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse_matrix)
```

```python
book_pivot.shape
```

```python
book_pivot.iloc[69:]
```

```python
distance, suggestions = model.kneighbors(book_pivot.iloc[69, :].values.reshape(1, -1), n_neighbors=6)
```

```python
### Testing the model

def recommend_books(book_name):
    index = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestions = model.kneighbors(book_pivot.iloc[index, :].values.reshape(1, -1), n_neighbors=6)

    books = book_pivot.index[suggestions[0]]
    for j in books:
        if(j == book_name):
            print(f"You searched '{book_name}'\n")
            print("Here are some similar books you might like:\n")
        else:
            print(j)
```

```python
book_name = "The Color of Water: A Black Man's Tribute to His White Mother"
recommend_books(book_name)
```
